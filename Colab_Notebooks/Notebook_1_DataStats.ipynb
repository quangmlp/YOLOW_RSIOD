# YOLO-IOD: Data Visualization
Notebook này dùng để vẽ các biểu đồ liên quan đến kích thước vật thể (Scale distribution) và đồng xuất hiện (Co-occurrence distribution) của các dataset DIOR và DOTA (dựa theo Figure 1 và Figure 5 của STAR-IOD).

```
import json
import matplotlib.pyplot as plt
import numpy as np
import os
from collections import defaultdict
import matplotlib.patches as patches

# Cấu hình font
plt.rcParams["font.family"] = "serif"
DATA_ROOT = "../data_drive"
```

## 1. Scale Distribution (Phân bổ kích thước)
Vẽ biểu đồ hình chữ nhật hiển thị kích thước tương đối của các bounding box (giống Figure 1a, 1b paper STAR-IOD).

```
def plot_scale_distribution(title, sizes, colors, labels):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    for size, color, label in zip(sizes, colors, labels):
        w, h = size
        rect = patches.Rectangle((50 - w/2, 50 - h/2), w, h, linewidth=2, edgecolor=color, facecolor="none", label=label)
        ax.add_patch(rect)
    ax.set_title(title, fontsize=14, fontweight="bold")
    ax.legend(loc="lower center", bbox_to_anchor=(0.5, -0.15), ncol=3)
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.savefig(f"{title.replace(' ', '_')}.png", dpi=300, bbox_inches="tight")
    plt.show()

# Giả lập số liệu scale của Ship và Plane (bạn có thể thay bằng hàm tính toán thực tế từ file json)
plot_scale_distribution("Scale Distribution of Ship", [(10, 20), (30, 60), (80, 90)], ["red", "blue", "yellow"], ["Small Object", "Medium Object", "Large Object"])
plot_scale_distribution("Scale Distribution of Plane", [(15, 15), (40, 40), (70, 70)], ["red", "blue", "yellow"], ["Small Object", "Medium Object", "Large Object"])
```

## 2. Co-occurrence Distribution
Biểu đồ tỷ lệ đồng xuất hiện của các lớp cũ và lớp mới (Giống Figure 5 paper STAR-IOD).

```
def plot_co_occurrence(datasets, old_only, new_only, co_occur):
    fig, ax = plt.subplots(figsize=(10, 4))
    y_pos = np.arange(len(datasets))
    
    p1 = ax.barh(y_pos, old_only, color="#f4a261", label="Only Old")
    p2 = ax.barh(y_pos, new_only, left=old_only, color="#2a9d8f", label="Only New")
    p3 = ax.barh(y_pos, co_occur, left=np.array(old_only)+np.array(new_only), color="#264653", label="Co-occurrence")
    
    ax.set_yticks(y_pos)
    ax.set_yticklabels(datasets)
    ax.set_xlabel("Percentage (%)")
    ax.set_title("Distribution of Old-New Category Co-occurrence")
    ax.legend(loc="upper center", bbox_to_anchor=(0.5, 1.15), ncol=3)
    
    # Thêm text
    for bars in [p1, p2, p3]:
        for bar in bars:
            width = bar.get_width()
            if width > 5:
                ax.text(bar.get_x() + width/2., bar.get_y() + bar.get_height()/2.,
                        f"{width}%",
                        ha="center", va="center", color="white", fontsize=10)
    plt.tight_layout()
    plt.savefig("Co_occurrence.png", dpi=300)
    plt.show()

# Số liệu tương đối từ paper
datasets = ["DIOR-IOD Train", "DIOR-IOD Test", "DOTA-IOD Train", "DOTA-IOD Test"]
old_only = [47.7, 46.1, 63.4, 60.6]
new_only = [40.1, 34.8, 15.6, 18.5]
co_occur = [12.2, 19.1, 21.0, 20.9]
plot_co_occurrence(datasets, old_only, new_only, co_occur)
```
